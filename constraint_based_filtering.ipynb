```python
# Sample item data with attributes (replace this with your data)
item_data = [
    {"item_id": 1, "name": "Movie A", "genre": "Action", "release_year": 1990},
    {"item_id": 2, "name": "Movie B", "genre": "Drama", "release_year": 2005},
    {"item_id": 3, "name": "Movie C", "genre": "Action", "release_year": 1995},
    {"item_id": 4, "name": "Movie D", "genre": "Comedy", "release_year": 2000},
    {"item_id": 5, "name": "Movie E", "genre": "Romance", "release_year": 2022},
    {"item_id": 6, "name": "Movie F", "genre": "Action,Comedy", "release_year": 2018},
    {"item_id": 7, "name": "Movie G", "genre": "RomCo", "release_year": 2019},
    {"item_id": 8, "name": "Movie H", "genre": "K-Drama", "release_year": 2015},
    {"item_id": 9, "name": "Movie I", "genre": "Action", "release_year": 2009},
    {"item_id": 10, "name": "Movie J", "genre": "Comedy", "release_year": 2010},
]
```

```python

# Define user constraints (e.g., genre and release year)
user_constraints = {
    "genre": "Action,Comedy"
}
```

```python

# Function to filter items based on constraints
def filter_items(items, constraints):
    filtered_items = []
    for item in items:
        if all(item[attr] == value for attr, value in constraints.items()):
            filtered_items.append(item)
    return filtered_items
```

```python

# Filter items based on user constraints
filtered_items = filter_items(item_data, user_constraints)
```

```python

# Sort the filtered items by a relevance score (you can define your scoring mechanism)
# In this example, we assume items are sorted by item_id (you can use a different scoring mechanism)
filtered_items.sort(key=lambda x: x["item_id"])
```

```python

# Print the filtered items
print("Filtered Items:")
for item in filtered_items:
    print(f"Item ID: {item['item_id']}, Name: {item['name']}")
```
